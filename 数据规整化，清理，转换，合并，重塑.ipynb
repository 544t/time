```python
#重排列和层次排序
frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                     columns=[['Ohio', 'Ohio', 'Colorado'],
                              ['Green', 'Red', 'Green']])
print(frame)
frame.index.names = ['key1','key2']
frame.columns.names = ['state','color']
#swapelevel()改变索引层次
print('\n---------源frame---------')
print(frame)
print('\n---------level = 1---------')
print(frame.sort_index(level = 1))
print('\n---------level = 1---------')
print(frame.sort_index(level = 0))

frame.swaplevel('key1','key2')#不知道有什么作用
```

```python
#按层级进行汇总统计
print(frame.sum(level = 'key2'))
print(frame.sum(level = 1))
#level = 'key2'等价于level = 1
print(frame.sum(level = 'color',axis = 1))#去掉axis = 1就报错，
# 因为level只能是Index,不能是column  
 
```

```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
print(df1)
print(df2)
print('\n--------merge()--------')
print(pd.merge(df1,df2))
print('\n--------外连接--------')
print(pd.merge(df1,df2,how = 'outer'))#外连接
print('\n--------左连接--------')
print(pd.merge(df1,df2,how = 'left'))#左连接
print('\n--------内连接--------')
print(pd.merge(df1,df2,on = 'key'))#指定聚合键
```

```python
#使用df的列进行索引(把某一列变成索引)
frame = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1),
                      'c': ['one', 'one', 'one', 'two', 'two',
                            'two', 'two'],
                      'd': [0, 1, 2, 0, 1, 2, 3]})
print(frame)
print(frame.set_index(['c','d']))
print(frame.reset_index())
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
print(pd.merge(df3,df4,left_on = 'lkey',right_on = 'rkey'))
# print(pd.merge(df3,df4,left_index = True))
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
print(left)
print(right)
print('\n-------------inner(交集)-------------')
print(pd.merge(left,right,on = ['key1','key2']))
print('\n-------------outer(补集)-------------')
print(pd.merge(left,right,on = ['key1','key2'],how = 'outer'))
print('------------suffixes打左右标签-------------')
print(pd.merge(left,right,on = 'key1'))
print(pd.merge(left,right,on = 'key1',suffixes = ['_left','_right']))
```

```python
#索引上的合并
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
print(left1)
print(right1)
print(pd.merge(left1,right1,left_on = 'key',right_index = True))
```

```python
#层次化索引
lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio',
                               'Nevada', 'Nevada'],
                      'key2': [2000, 2001, 2002, 2001, 2002],
                      'data': np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['Nevada', 'Nevada', 'Ohio', 'Ohio',
                              'Ohio', 'Ohio'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['event1', 'event2'])
print(lefth)
print(righth)
#以列表的形式
print(pd.merge(lefth,righth,left_on = ['key1','key2'],right_index = True))

print('------------同时使用双方的索引------------')
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
print(pd.merge(left2,right2,left_index = True,right_index = True))
```

```python
#join实例方法
print(left2.join(right2,how = 'outer'))

another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                       index=['a', 'c', 'e', 'f'],
                       columns=['New York', 'Oregon'])
print(another)
print('\n--------三表连接-------')
print(left2.join([right2,another]))#注意这里为中括号
print(left2.join([right2,another],how = 'outer'))
```

```python
#轴向连接（连接，绑定，堆叠）
#numpy数组
print('------numpy数组-----')
arr = np.arange(12).reshape(3,4)
print(arr)
print(np.concatenate([arr,arr],axis = 1))#横向拼接
print(np.concatenate([arr,arr]))#纵向堆叠

#series
print('\n--------series--------')
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
print(pd.concat([s1,s2,s3]))#横向
print(pd.concat([s1,s2,s3],axis = 1))#纵向
```

```python
s4 = pd.concat([s1*5,s3])
print('-------s4------')
print(s4)
print('------s1------')
print(s1)
print(pd.concat([s1,s4],axis = 1))#默认为‘outer’
print(pd.concat([s1,s4],axis = 1,join = 'inner'))#注意这里是join不是how
print(pd.concat([s1,s4],axis = 1,join = 'inner',
               join_axes = [['a','b','c','d']]))#注意这里是两个方括号
#join_axes只能给一个参数，这里这个参数是列表['a','b','c','d']
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
print(s1)
print(s2)
print(s3)
#堆叠时增加层次化索引
print('\n-----------堆叠时增加层次化索引----------')
r = pd.concat([s1,s2,s3],keys = ['one','two','three'])
print(r)
#沿着axis = 0对series进行堆叠，keys就会成为series的索引
#沿着axis = 1对series进行堆叠，keys就会成为DF的表头
print(pd.concat([s1,s2,s3],axis = 1,keys = ['one','two','three']))
```

```python
#unstack函数
print(r.unstack())
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                   columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                   columns=['three', 'four'])
print(df1)
print(df2)
print(pd.concat([df1,df2],axis = 1,keys = ['level1','level2']))
#可传入一个字典,主键就是keys的值
print(pd.concat({'level1':df1,'level2':df2},axis = 1))
#为层次化索引命名
pd.concat([df1,df2],axis = 1,keys = ['level1','level2'],
         names = ['upper','lower'])
```

```python
#合并重叠数据
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
              index=['f', 'e', 'd', 'c', 'b', 'a'])
print(a)
print(b)
b[-1] = np.nan
#使用np.where()
print(np.where(pd.isnull(a),b,a))
#使用combine_first()自己有的用自己的，没有的combine_first别人的
#就是打补丁
print('--------combine_first---------')
print(b[:-2])
print(a[2:])
print('\n-----------------')
print(b[:-2].combine_first(a[2:]))
```

```python
#重塑（reshape）和轴向旋转(pivot)
# stack（将横着的转为竖着的）
# unstack（将竖着的转为横着的）

data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                 index=pd.Index(['Ohio', 'Colorado'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],
                    name='number'))
print(data)
print('------------------')
print(data.stack())
print(data.stack().unstack())
```

```python
import pandas as pd
import numpy as np
print(data.stack().unstack(0))#0为层级，-1为最内层，0为最外层，默认为-1
print(data.stack().unstack('state'))#可直接指定层级名称
```

```python
s1 = pd.Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([4, 5, 6], index=['c', 'd', 'e'])
data2 = pd.concat([s1, s2], keys=['one', 'two'])
print(data2)
print(data2.unstack())
print(data2.unstack().stack())#可逆的
print(data2.unstack().stack(dropna = False))#dropna默认为true，
# 表示自动去除null值，dropna = False为保留null值
```

```python
result = data.stack()
df = pd.DataFrame({'left': result, 'right': result + 5},
                  columns=pd.Index(['left', 'right'], name='side'))
print(df)
print('\n--------------------')
print(df.unstack('state').stack('side'))
```

```python
# 将长透视为宽
data = pd.read_csv(r'C:\Users\123\Desktop\macrodata.csv')
print(data.head())
periods = pd.PeriodIndex(year = data.year,quarter = data.quarter,
                     name = 'data')
print('\n-------------------------')
print(periods)
print('\n------------------------')
columns = pd.Index(['realgdp','infl','unemp'],name = 'item')
data = data.reindex(columns = columns)
data.index = periods.to_timestamp('D','end')
print(data.head())
ldata = data.stack().reset_index().rename(columns = {0:'value'})
print(ldata.head())
```

```python
#pivot方法
pivoted = ldata.pivot('data','item','value')
print(pivoted.head())
#data对应index,value对应columns
```

```python
ldata['value2'] = np.random.randn(len(ldata))
print(ldata.head())
print('-------------目标------------')
pivoted = ldata.pivot('data','item')#横轴data,纵轴item
print(pivoted.head())
print(pivoted['value'][:5])
print(pivoted['value2'][:5])

print('-----------------------')
unstack = ldata.set_index(['data','item']).unstack('item')
print(unstack.head())
```

```python
#将宽透视为长
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
print(df)
#id_vars指定列进行unpivot操作(有id的意思)
# id_vars对应index，value_vars对应columns!!!!!
melted = pd.melt(df,id_vars='key')
print(melted)
print(melted.pivot('key','variable','value'))
pd.melt(df,value_vars=['A','B','C'])
```
