```python
#缺失值处理
# dropna fillna isnull notnull
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1] = np.nan
df.iloc[:2,2] = np.nan
print(df)
print(df.dropna())
print('-------至少有两个非null值-------')
print(df.dropna(thresh=2))
```

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
print(data)
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}
l = data['food'].str.lower()
print(l)
```

```python
data['animal'] = l.map(meat_to_animal)
print(data['animal'])
print(data)
print(' \n---------这个需要注意！--------')
print(data['food'].map(lambda x: meat_to_animal[x.lower()]))
```

```python
#分箱
#按给定数值分箱
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18,25,35,60,100]
cats = pd.cut(ages,bins)#默认参数right=True，表示右闭
#参数labels=None，设置每个箱子的名称
print(cats)
print(pd.value_counts(cats))#分箱并计数
name = ['青年','老青年','中年','老年']
cats = pd.cut(ages,bins,labels = name)
print(cats)
print(cats.value_counts())

#给定分箱数（按最大值最小值分成X分）
cats2 = pd.cut(ages,4,precision=2)#precision=3为默认小数位数
print(cats2)
```

```python
#qcut平均分箱（各箱分到的数量相等）
data = np.random.randn(1000)
q = pd.qcut(data,4)
print(q)
print(q.value_counts())

#自定义分位数进行分箱
print('-------------------------')
qq = pd.qcut(data,[0,0.3,0.6,0.8,1])
print(qq)
print(qq.value_counts())
```

```python
#数据筛选（head,describe,索引）
data = pd.DataFrame(np.random.randn(1000, 4))
print(data.head())
print(data.describe())
print('\n-------绝对值超过3--------')
print(data[2][np.abs(data[2]>2.5)])
print(data[np.abs(data>2.9).any(1)])#一行有一个大于2.9

#np.sign()>0为1，<0为-1，=0为0
print('\n----------np.sign()----------')
print(np.sign(data).head())
```

```python
#随机抽样
df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
print(df)
s = np.random.permutation(5)#取源数据中的X个数据进行随机排序
print(s)
print(df.sample(n=3))
print('\n--------take--------')
print(df.take(s))
print('\n--------iloc--------')
print(df.iloc[[2,1,0,4,3]])
#df.take(s)等价于df.iloc[[2,1,0,4,3]
```

```python
#计算指标（）get_dummies
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                   'data1': range(6)})
print(df)
print(pd.get_dummies(df['key']))
d = pd.get_dummies(df['key'],prefix = 'key')
print(d)
# du = df['data1'].join(d) 
# 上面式子出错 AttributeError: 'Series' object has no attribute 'join'
#df['data1']是series，所以没有join方法
print('\n-----------join------------')
du = df[['data1']].join(d)
print(du)
```

```python
#计算指标（）get_dummies应用
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(r'C:\Users\123\Desktop/movies.dat', sep='::',
                       header=None, names=mnames,engine='python')
print(movies[:10])
all_genres = []
for i in movies['genres']:
    all_genres.extend(i.split('|'))
genres = pd.unique(all_genres) 
print('\n-------genres列表--------\n')
print(genres)
print(type(genres))
```

```python
#构建全零矩阵
zeros_matrix = np.zeros((len(movies),len(genres)))#注意这里
# np.zeros(())两层括号
print(zeros_matrix)
dummies = pd.DataFrame(zeros_matrix,columns = genres)
print(dummies.head(2))
```

```python
gen = movies['genres'][0]
print(gen)
a = gen.split('|')
print(a) 
dummies.columns.get_indexer(a)
```

```python
for i, gen in enumerate(movies['genres']):
    indices = dummies.columns.get_indexer(a)
    dummies.iloc[i,indices] = 1
    
movies_windic = movies.join(dummies.add_prefix('Genre_'))   
movies_windic.head()
```

```python
# 练习一下
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(r'C:\Users\123\Desktop/movies.dat', sep='::',
                       header=None, names=mnames,engine='python')
print(movies[:3])
genres = []
for i in movies['genres']:
    genres.extend(i.split('|'))
    
genres_list = pd.unique(genres)
print(genres_list)  
movies_matrix = np.zeros((len(movies),len(genres_list)))
movies_matrix2 = pd.DataFrame(movies_matrix,columns = genres_list)
print(movies_matrix2.head(2))

for i, gen in enumerate(movies['genres']):
    indices = movies_matrix2.columns.get_indexer(gen.split('|'))
    movies_matrix2.iloc[i,indices] = 1

end = movies.join((movies_matrix2))
end.head()

#1,通过movies表genres列提取columnsX（去重）
#2,构造0矩阵，列名设为columnsX
#3,利用enumerate(),iloc(),把每一行相应的列的值取为1
#4,原表join(改造后的0矩阵表)
#5,成功！
```

```python
#正则表达式
# re模块
import re
t = 'a    d   \nc'
s = re.split('\s+',t)
print(s)
# re.complie()可生成一个可复用对象
mothed = re.compile('\s+')
a = re.split(mothed,t)
# 等价于
b = mothed.split(t)
print(a,b)
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
red = re.compile(pattern,flags=re.IGNORECASE)
red.findall(text)
#又是版本问题
```
