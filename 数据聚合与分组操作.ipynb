第十章：数据聚合与分组操作


```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
print(df)
a = df['data1'].groupby(df['key1'])
print(a)
print(a.sum())
b = df['data2'].groupby(df['key2'])
print(b)
print(b.sum())
#多个分组键
c = df['data1'].groupby([df['key1'],df['key2']]).sum()
#注意groupby里面两个方括号，括起来表示作为列表传给第一个参数by，
# 不括起来就默认传一个给第二个参数axis，就会报错
print(c)
print(c.unstack())#stack是‘叠起来的意思’，所以表示行转列，
# unstack是‘把叠起来的再铺开’,所以是列转行
```

```python
#分组键除了Series还可以是数组
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
years = np.array([2005, 2005, 2006, 2005, 2006])
print(df['data1'].groupby([states,years]).sum())
df.groupby(['key1','key2']).size()#显示符合索引的数值个数
```

```python
#遍历各分组
for a, b in df.groupby('key1'):
    print(a)
    print(b)

print('\n-----------------') 

for (k1,k2), b in df.groupby(['key1','key2']):
    print((k1,k2))
    print(b)
#每组的标题即为分组键
```

```python
# 用字典也可以有类似提取分组的效果
p = dict(list(df.groupby('key1')))
print(p['b'])
```

```python
print(df.dtypes)#注意，这里为啥没有括号？？？？
g = df.groupby(df.dtypes,axis = 1)
print(g)
for x, y in df.groupby(df.dtypes,axis = 1):
    print(x)
    print(y)
```

```python
#选取一个或者多个列
df.groupby('key1')['data1']#放一列
df.groupby('key1')[['data1','data2']]#放两列
# 等价于
df['data1'].groupby(df['key1'])
df[['data1','data2']].groupby(df['key1'])#注意这里和上面的区别，这里加了df
# !!!!!!
print(df.groupby(['key1','key2'])['data1'].mean())
print('--------------------------')
print(df['data1'].groupby([df['key1'],df['key2']]).mean())
```

```python
#通过字典或series分组?????????????????????????/
people = pd.DataFrame(np.random.randn(5, 5),
                      columns=['a', 'b', 'c', 'd', 'e'],
                      index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
print(people)
#添加几个nan值
people.iloc[2:3,[1,2]] = np.nan
print(people)
print('\n------------通过字典分组----------------')
mapping = {'a': 'red', 'b': 'red', 'c': 'blue',
           'd': 'blue', 'e': 'red', 'f' : 'orange'}
g = people.groupby(mapping)
print(g.sum())
h = people.groupby(mapping,axis = 1)
print(h.sum())
```

```python
#series
m = pd.Series(mapping)
print(m)
people.groupby(m,axis = 1).sum()
```

```python
#通过函数进行分组
print(people.groupby(len).sum())
key_list = ['one', 'one', 'one', 'two', 'two']
print(people.groupby([len,key_list]).sum())            
```

```python
#根据索引级别分组
columns = pd.MultiIndex.from_arrays([['US', 'US', 'US', 'JP', 'JP'],
                                    [1, 3, 5, 1, 3]],
                                    names=['cty', 'tenor'])
df = pd.DataFrame(np.random.randn(4, 5), columns=columns)
print(df)
print(df.groupby(level = 'cty',axis = 1).count())
# print(df.groupby(level = 'cty').count())报错！！原因不明
```

```python
#数据聚合
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
print(df)
print(df.groupby('key1')['data1'].quantile(0.75))
print(df.groupby('key1')['data1'].quantile(0.5))
print(df.groupby('key1')['data1'].quantile(0.25))#这个分位机制不理解？？？？
```

```python
#agg与aggregate
def pek_pek(arr):
    return arr.max() - arr.mean()
print(df.groupby('key1').agg(pek_pek))

#describe()
print(df.groupby('key1').describe())
```

```python
tips = pd.read_csv(r'C:\Users\123\Desktop\tips.csv')
print(tips.head())
#添加消费占比列
print('\n------消费占比列------')
tips['tip_pct'] = tips['tip']/tips['total_bill']
print(tips.head())
g_pct = tips.groupby(['day','smoker'])['tip_pct'].agg('mean')
print(g_pct)
print('---------使用多个函数---------')
f_pct = tips.groupby(['day','smoker'])['tip_pct'].agg(['mean','std',pek_pek])
print(f_pct)
```

```python
#（name,func）
tt = tips['tip_pct'].groupby([tips['day'],tips['smoker']]).agg([('aaa','mean'),('bbb','sum')])
print(tt)
```

```python
func = ['mean','max','sum']
ff = tips[['tip_pct','total_bill']].groupby([tips['day'],tips['smoker']]).agg(func)
print(ff.head())
print(ff['tip_pct'].head())
```

```python
#对不同的列使用不同的函数
gd = tips.groupby(['day','smoker'])
fc = {'tip':'max','size':'sum'}
q = gd['tip_pct','total_bill'].agg(fc)
print(q)
print('\n----------有索引状态-------------')
#无索引状态
bb = tips.groupby(['day','smoker'],as_index = True).mean()
qq = tips.groupby(['day','smoker'],as_index = False).mean()
print(bb)
print('\n----------无索引状态-------------')
print(qq)
```

```python
# transform()
key = ['one','two','one','two','one']
print(key)
print(people.groupby(key).mean())
print(people.groupby(key).transform('mean'))
#transform()不会对对列进行聚合，保持与原表相同的行数，便于与原表进行连接运算等操作
def demean(arr):
    return arr - arr.mean()
d = people.groupby(key).transform(demean)
print(d)#注意，为什么有null值？？？？
# print(d.groupby(key).mean())这个值不对？？？？
```

```python
#apply
def top(df,n = 5,columns = 'tip_pct'):
    return df.sort_index(by = columns)[-n:]
print(top(tips,n = 6))
print(tips.groupby('smoker').apply(top))
#apply里可以修改参数
tips.groupby(['smoker','day']).apply(top,n=1,columns='total_bill')
```

```python
r = tips.groupby('smoker')['tip_pct'].describe()
print(r)
print(r.unstack())#展开，有时候不是列转行，而是数据维度意义上的展开
```

```python
#禁止分组键
print(tips.groupby('smoker').apply(top))
print('\n-------不要分组键--------')
print(tips.groupby('smoker',group_keys = False).apply(top))
```

```python
#分位数和桶分析（cut,qcut）
frame = pd.DataFrame({'data1': np.random.randn(1000),
                      'data2': np.random.randn(1000)})
quartiles = pd.cut(frame.data1, 4)
print(quartiles.head())

def get_stas(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}

grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stas).unstack()#这里要注意！！！！！
```

```python
#qcut()
quartiles = pd.qcut(frame.data1, 10)
grouped = frame.data2.groupby(quartiles)
print(grouped.apply(get_stas).unstack())
# #不要data1索引
# quartiles = pd.qcut(frame.data1, 10,labels = False)
# print(grouped.apply(get_stas).unstack())
#好像没啥区别
```
